## What is Matplotlib?

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

```python
import matplotlib as plt

print('Current version of Matplotlib :',plt.__version__)
```

## Pyplot

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Draw a line from points

xpoints = np.array([0,10])
ypoints = np.array([0,100])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
# plot multiple points

# Draw a line from points

xpoints = np.array([0,0,10,30,50,70,100])
ypoints = np.array([0,30,5,50,30,90,100])

plt.plot(xpoints,ypoints)
plt.show()
```

## Default X-Points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 (etc. depending on the length of the y-points.)

```python
ypoints = np.array([1,9,2,8,3,7,4,6,5])

plt.plot(ypoints)
plt.show()
```

## Markers

```python
# Draw a line from points

xpoints = np.array([0,0,10,30,50,70,100])
ypoints = np.array([0,30,5,50,30,90,100])

plt.plot(xpoints,ypoints,'*r')
plt.show()
```

```python
ypoints = np.array([1,9,2,8,3,7,4,6,5])
plt.plot(ypoints, '*-.')
plt.show()
```

## Marker size

```python
ypoints = np.array([1,9,2,8,3,7,4,6,5])
plt.plot(ypoints, '*-.', ms = 20)
plt.show()
```

## Marker colour

```python
ypoints = np.array([1,9,2,8,3,7,4,6,5])
plt.plot(ypoints, '*-.',ms = 20 ,mec = 'k')
plt.show()
```

```python
ypoints = np.array([1,9,2,8,3,7,4,6,5])
plt.plot(ypoints, '*-.',ms = 20 ,mfc = 'k')
plt.show()
```

## Linestyle

## Linewidth

## Title

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)

plt.title('Title')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# position the title
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)

plt.title('Title', loc= 'right')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# position the title
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)

plt.title('Title', loc= 'right')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()

plt.show()
```

```python
#  axis wise grid
# position the title
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)

plt.title('Title', loc= 'right')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(axis='x')
# plt.grid(axis='y')

plt.show()
```

## Title on subplots

```python
# Plot 1

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Income')
# Plot 2

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.title('Sales')
plt.suptitle('Nike')
plt.show()
```

## Display multiple plots:- Subplot

The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

```python
# Plot 1

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1,2,1)
plt.plot(x,y)

# Plot 2

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
# Plot 1

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2,1,1)
plt.plot(x,y)

# Plot 2

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,1,2)
plt.plot(x,y)
plt.show()
```

## Colormap

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

## Size

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

## Alpha

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

## Combine Color Size and Alpha

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

## Labels

```python
x = np.array([10,20,30,40])
y = np.array(['Apple','Banana','Mango','Orange'])

plt.pie(x,labels = y)
plt.show()
```

## Explode

```python
x = np.array([10,20,30,40])
mylabel = np.array(['Apple','Banana','Mango','Orange'])
myexplode = np.array([0.5,0,0,0])

plt.pie(x, labels = mylabel, explode = myexplode)
plt.show()
```

```python
# shadow
x = np.array([10,20,30,40])
mylabel = np.array(['Apple','Banana','Mango','Orange'])
myexplode = np.array([0.2,0,0,0])

plt.pie(x, labels = mylabel, explode = myexplode, shadow = True)
plt.show()
```
